```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Gender: {self.gender}")


class Employee(Person):
    def __init__(self, name, age, gender, employee_id, salary):
        super().__init__(name, age, gender)
        self.employee_id = employee_id
        self.salary = salary

    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.employee_id}, Salary: ${self.salary}")


class Doctor(Employee):
    def __init__(self, name, age, gender, employee_id, salary, specialization):
        super().__init__(name, age, gender, employee_id, salary)
        self.specialization = specialization

    def display_info(self):
        super().display_info()
        print(f"Specialization: {self.specialization}")
        

class Nurse(Employee):
    def __init__(self, name, age, gender, employee_id, salary, department):
        super().__init__(name, age, gender, employee_id, salary)
        self.department = department

    def display_info(self):
        super().display_info()
        print(f"Department: {self.department}")


if __name__ == "__main__":
    doctor = Doctor("Dr. Smith", 35, "Male", "DOC123", 80000, "Cardiology")
    nurse = Nurse("Nurse Johnson", 28, "Female", "NUR456", 50000, "ICU")

    print("Doctor Information:")
    doctor.display_info()

    print("\nNurse Information:")
    nurse.display_info()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


class Patient(Person):
    def __init__(self, name, age, patient_id):
        super().__init__(name, age)
        self.patient_id = patient_id

    def display(self):
        super().display()
        print("Patient ID:", self.patient_id)


class Doctor(Person):
    def __init__(self, name, age, doctor_id, specialty):
        super().__init__(name, age)
        self.doctor_id = doctor_id
        self.specialty = specialty

    def display(self):
        super().display()
        print("Doctor ID:", self.doctor_id)
        print("Specialty:", self.specialty)


class Surgeon(Doctor):
    def __init__(self, name, age, doctor_id, specialty, procedures):
        super().__init__(name, age, doctor_id, specialty)
        self.procedures = procedures

    def display(self):
        super().display()
        print("Procedures:", ', '.join(self.procedures))


class Nurse(Person):
    def __init__(self, name, age, nurse_id):
        super().__init__(name, age)
        self.nurse_id = nurse_id

    def display(self):
        super().display()
        print("Nurse ID:", self.nurse_id)


# Usage
patient = Patient("John Doe", 30, "P123")
doctor = Doctor("Dr. Smith", 45, "D567", "Cardiologist")
surgeon = Surgeon("Dr. Johnson", 50, "D789", "Surgeon", ["Appendectomy", "Laparoscopy"])
nurse = Nurse("Jane Smith", 28, "N456")

patient.display()
print("\n")
doctor.display()
print("\n")
surgeon.display()
print("\n")
nurse.display()
```
